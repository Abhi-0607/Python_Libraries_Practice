```python
import tensorflow as tf
from tensorflow import keras
import matplotlib.pyplot as plt
from tensorflow.keras import layers
import numpy as np
```

```python
x = np.array([1,2,3,4,5,6,7,8], dtype=float)
y = np.array([0,0,0,1,1,1,1,1], dtype=float)
```

```python
model = keras.Sequential([layers.Input(shape=(1,)),
                          layers.Dense(8, activation="relu"),
                          layers.Dense(4, activation="relu"),
                          layers.Dense(1, activation="sigmoid")])
```

```python
model.compile(optimizer = "adam", loss='binary_crossentropy', metrics=['accuracy'])
```

```python
model.summary()
```

```python
history = model.fit(x,y, epochs = 50, batch_size=2)
```

```python
plt.plot(history.history['loss'])
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Training Loss")
plt.title("Training Loss")
plt.show()
```

```python
test_hours = np.array([2,5,7], dtype=float)
predictions = model.predict(test_hours)
print(predictions)
```

```python
print((predictions > 0.5).astype(int))
```

```python
plt.plot(history.history['accuracy'])
plt.plot(history.history['loss'])
plt.legend(['Accuracy','Loss'])
plt.show()
```
