```python
#Reverse a string / pallindrome or not
word = input("Enter a word to test if it is a pallindrome or not:")
if(word == word[::-1]):
    print("It is a pallindrome")
    print(word)
else:
    print("Not a pallindrome")
    print(word[::-1])
```
