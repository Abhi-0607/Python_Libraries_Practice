```python
#importing numpy
#converting list to arrays in python
import numpy as np

list = [1,2,3,4,5]
print(list)
arr = np.array(list)
print(arr)
```

```python
#creating a matrix using numpy, and defining operations of matrix
import numpy as np

matrix = np.array([[1,2,3] , [4,5,6]])
print(matrix)
print("Size of the above matrix is:",matrix.shape)
print("Data type of the matrix is:",matrix.dtype)
```

```python
#creating an array of zeroes and ones

zeros = np.zeros((3,8))
print("Matrix of Zeroes:")
print(zeros)

print("Matrix of Ones:")
ones = np.ones((4,8))
print(ones)
```

```python
#matix with range of numbers
import numpy as np

#arange operation
range_arr = np.arange(0,10,3) #prints the first number, and futher steps in the given number and prints within the range
print(range_arr)

#linspace operation
range_arr = np.linspace(0,10,3) #gives the parts of the given range. ex: number line
print(range_arr)
```

```python
# arranging arrray to matrix with "reshape" operation and back to 1d using "flatten" operaion
import numpy as np

#reshape
print("Reshape operation:")
arr = np.arange(12).reshape(4,3)
print(arr)

#flatten
print("\nFlatten operation:")
flat = arr.flatten()
print(flat)
```

```python
#Creating array and performing array Operations: min, max, sum, slicing
import numpy as np

arr = np.array([1,2,3,3,3,4,5])
print("Array:",arr)

print("Minimum value:",arr.min()) #min value
print("Maximum Value:",arr.max()) #max value
print("Sum of values in array:",arr.sum()) #sum 
print("Sliced array by 3:",arr[:3]) #slicing array
```

```python
#Creating ana array and finding mean, median and standard deviation
import numpy as np

arr = np.array([1,2,3,4,5])
print("Mean:",np.mean(arr))
print("Median:",np.median(arr))
print("Standard Deviation:",np.std(arr))
```

```python
#Creating a 2D aray and performing few operations: slicing, accessing specific element
import numpy as np

matrix = np.array([[1,2,3,],[4,5,6]])
print("Matrix:")
print(matrix)

#slicing
print("\nSliced Matrix:")
print(matrix[0])

#specific element
print("\nSpecific Element:")
print(matrix[1,2])
```

```python
# Array Operations with other arrays\
import numpy as np

arr1 = np.array([1,3,5,7,9,11])
arr2 = np.array([2,4,6,8,10,12])

#performing element-wise addition
print(arr1 + arr2)

#performing element-wise muliplication
print(arr1 * arr2)
```

```python
# Matrix Operations: multiplication
import numpy as np

mat1 = np.array([[1,2,3],[6,7,8]])
mat2 = np.array([[2,4,6],[1,2,3],[1,2,3]])
print("Matrix 1:\n",mat1)
print("\nMatrix 2:\n",mat2)

#Matrix Multiplication(Dot)
print("\nDot Product:\n",np.dot(mat1, mat2))

#Transpose
print("\nTranspose:\n",mat1.T)
```

```python
#More array Operation: unique, replcing
import numpy as np

arr = np.array([1,2,3,3,3,4,5])
print("Array:",arr)

#finding unique elemnts only
print("Uniques:",np.unique(arr))

#replacing elements based on condition
arr[arr==3]=0
print("Replced by 0 if element is 3:",arr)


```
