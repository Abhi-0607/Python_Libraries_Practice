```python
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

data = { 'Size_sqt': [1000, 1500, 1200, 1800, 1100], 'Price': [300,450,500,800,330]}
df = pd.DataFrame(data)
x = df[['Size_sqt']]
y = df['Price']

#Split into training and testing data
x_train, x_test, y_train, y_test = train_test_split(x,y, test_size = 0.2, random_state = 42)

#creating and training Linear model
model = LinearRegression()
model.fit(x_train, y_train) #fit the model to training data



```

```python
#Intercept learned parameters

slope_m = model.coef_[0]
intercept_b = model.intercept_
print(f"Learned Slope (m): {slope_m: .2f}")
print(f"Learned Intercept (b): {intercept_b: .2f}")
```

```python
# predfict on test set and evaluate using r square

y_pred = model.predict(x_test)
r_sq = model.score(x_test, y_test)
print(f"R-Squared Score on Test Set: {r_sq: 4f}")
```

```python
#prediction for new unseen example

new_house_size = np.array([[11300]])
predicted_price = model.predict(new_house_size)
print(f"Predicted Price for 1300 sqft: ${predicted_price[0]: 2f}k")
```

```python
#data visualization

plt.scatter(x, y, label = "Actual Data")
plt.plot(x, model.predict(x), color = 'red', label = 'Regression Line')
plt.xlabel("Size (sqt)")
plt.ylabel('Price')
plt.title('Simple Linear Regression')
plt.grid(True)
plt.show()
```
