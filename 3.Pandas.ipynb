```python
#creatinf two dataframes and their operartion
df1 = pd.DataFrame({'Roll.No': ['22321A0501','22321A0502', '22321A0503'], 'Name': ['Abhignya','Abhinaya','Aditi']})
df2 = pd.DataFrame({'Name':['Abhignya','Abhinaya','Aditi'],'Hobbies': ['Boxing','Drawing','Reading']})
print(df1)
print("\n")
print(df2)
```

```python
#two dataframe operations

#concating them
df3 = pd.concat([df1,df2], axis = 0)
df3
```

```python
#mergign them
merged = pd.merge(df1,df2, on = 'Name')
print(merged)
```
