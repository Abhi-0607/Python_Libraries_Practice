```python
import matplotlib.pyplot as plt

x = [5,6,8]
y = [3,4,9]
plt.plot(x,y, marker='*')
plt.title("Boss, it is a line graph") #title for the graph
plt.show()
```

```python
x = [5,6,8]
y = [3,4,9]
plt.plot(x,y, marker='*')
plt.title("Boss, it is a line graph") #title for the graph
plt.xlabel("X value") # x title
plt.ylabel("Y value") # y title
plt.show()
```

```python
x = [5,6,8]
y = [3,4,9]
z = [10,20,30]
plt.plot(x,y, marker = '*')
plt.plot(x,z, marker = '*')
plt.title("This is a double graph")
plt.xlabel("X Values")
plt.ylabel("Y Value")
plt.show()
```

```python
# creating a abr graph
months = ["Jan", "Feb", "March", "April" , "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
sales = [100,200,300,500,600,100,700,800,100,200,900,500]
plt.bar(months, sales)
plt.title("Sales of year 2025 - monthwise")
plt.xlabel("Months")
plt.ylabel("Sales")
```
